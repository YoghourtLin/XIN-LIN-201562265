```python
# vusualisation for the pie chart about percentage of each big label in the whole dataset

All_comments = 12293

num_comments = [541,4398,226,539,956,3381,184,366,1702]

labels = ['Access', 'Care received', 'Dignity', 'Environment', 'Communication', 'Staff', 'Transition/Coordination', 'Miscellaneous', "Couldn't be improved"]

colors = ['green','yellow','red','blue','purple','pink','white', 'cyan','orange']

sizes = [num_comments[0]/All_comments*100,num_comments[1]/All_comments*100,num_comments[2]/All_comments*100,num_comments[3]/All_comments*100,
         num_comments[4]/All_comments*100,num_comments[5]/All_comments*100,num_comments[6]/All_comments*100,num_comments[7]/All_comments*100,num_comments[8]/All_comments*100]

plt.pie(sizes,labels=labels,shadow=True,colors=colors,autopct='%.2f%%',radius=2.5)
plt.show()
```
