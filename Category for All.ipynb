```python
# This ipynb is for summarising big labels and categoryies in the dataset

import pandas as pd
```

```python
# Read csv file
dataset  = pd.read_csv('All_Data_Original.csv')
```

```python
dataset
```

```python
# Drop useless columns

dataset = dataset.drop(columns = 'feedback', inplace=False)
dataset = dataset.drop(columns = 'space', inplace=False)
dataset = dataset.drop(columns = 'sentiment_score', inplace=False)
dataset = dataset.drop(columns = 'sentiment tag', inplace=False)
```

```python
dataset
```

```python
# Eliminate duplicate rows

dataset = dataset.drop_duplicates(keep = 'first')
```
